```python
import re
import string
```

```python
def textProcessing(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove text within square brackets
    text = re.sub(r'\[.*?\]', '', text)
    
    # Replace non-word characters (excluding spaces) with a space
    text = re.sub(r'\W', ' ', text)
    
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # Remove punctuation
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    
    # Remove numeric characters
    text = re.sub(r'\d+', '', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
news['text'] = news['text'].apply(textProcessing)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
def output_label(n):
    if n==0:
        return "Its a Fake News"
    elif n==1:
        return "Its a Real News"
```

```python
def manual_testing(news):
    # Create a DataFrame from the input news text
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    
    # Apply the wordopt function to preprocess the text
    new_def_test["text"] = new_def_test["text"].apply(textProcessing)
    
    # Vectorize the preprocessed text
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test) 
    
    # Make predictions using the models
    pred_lr = LR.predict(new_xv_test)
    
    # Print each prediction on a separate line
    print("according to our AI model: ")
    print(output_label(pred_lr[0]))


```
